```python
#basic processing-lower,url,emoji,etc
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv("C:/Users/Rohan/Downloads/SAP/Codes/Adityas Tweets/Adityas Tweets Uncleaned.csv")
```

```python
df
```

```python
df.drop('Unnamed: 0',inplace=True,axis=1)
df
```

```python
df['Tweets']=df['Tweets'].str.lower()
df
```

```python
#removing URLs
import re
def remove_URL(tweet):
    return re.sub(r"http\S+", " ", tweet) 
```

```python
df['Tweets']=df['Tweets'].apply(remove_URL)
df
```

```python
#removing newline
import re
def remove_newline(tweet):
    return re.sub("\n", " ", tweet)
```

```python
df['Tweets']=df['Tweets'].apply(remove_newline)
df
```

```python
#remove emojis and emoticons
import re
import pickle
from emot.emo_unicode import UNICODE_EMOJI # For emojis
from emot.emo_unicode import EMOTICONS_EMO # For EMOTICONS
```

```python
def remove_emoji(string):
    emoji_pattern = re.compile("["
                               u"\U0001F600-\U0001F64F"  # emoticons
                               u"\U0001F300-\U0001F5FF"  # symbols & pictographs
                               u"\U0001F680-\U0001F6FF"  # transport & map symbols
                               u"\U0001F1E0-\U0001F1FF"  # flags (iOS)
                               u"\U00002500-\U00002BEF"  # chinese char
                               u"\U00002702-\U000027B0"
                               u"\U00002702-\U000027B0"
                               u"\U000024C2-\U0001F251"
                               u"\U0001f926-\U0001f937"
                               u"\U00010000-\U0010ffff"
                               u"\u2640-\u2642"
                               u"\u2600-\u2B55"
                               u"\u200d"
                               u"\u23cf"
                               u"\u23e9"
                               u"\u231a"
                               u"\ufe0f"  # dingbats
                               u"\u3030"
                               "]+", flags=re.UNICODE)
    return emoji_pattern.sub(r'', string)
```

```python
df['Tweets']=df['Tweets'].apply(remove_emoji)
df
```

```python
with open('Emoticon_Dict.p', 'rb') as fp:
    Emoticon_Dict = pickle.load(fp)

```

```python
def remove_emoticons(text):
    emoticon_pattern = re.compile(u'(' + u'|'.join(k for k in Emoticon_Dict) + u')')
    return emoticon_pattern.sub(r'', text)

```

```python
df['Tweets']=df['Tweets'].apply(remove_emoticons)
df
```

```python
#trimming
df['Tweets']=df['Tweets'].str.strip()
df
```

```python
#remove additional 
def remove_extra_whitespace(tweet):
    return re.sub("\s\s+", " ", tweet)
```

```python
df['Tweets']=df['Tweets'].apply(remove_extra_whitespace)
df
```

```python
#save
df.to_csv('Adityas_Basic_Processed.csv')
```
