```python
import pandas as pd
df=pd.read_csv('C:/Users/Rohan/Downloads/SAP/Codes/Text2Emotion/TextBlob Uncleaned.csv')
df
```

```python
pos=[]
neg=[]
neu=[]
for i in df['Polarity']:
    if i>(0.2):
        pos.append(i)
    elif i<(-0.2):
        neg.append(i)
    else:
        neu.append(i)
posa=[abs(number) for number in pos]
nega=[abs(number) for number in neg]
neua=[abs(number) for number in neu]
print(sum(posa))
print(sum(nega))
print(sum(neua))
print(len(posa))
print(len(nega))
print(len(neua))
```
