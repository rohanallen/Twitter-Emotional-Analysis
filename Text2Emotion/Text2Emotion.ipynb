```python
import pandas as pd
df=pd.read_csv('C:/Users/Rohan/Downloads/SAP/Codes/Adityas Tweets/Adityas Tweets Uncleaned.csv')
df
```

```python
from textblob import TextBlob
```

```python
def getPolarity(tweet):
   return TextBlob(tweet).sentiment.polarity
```

```python
def getSubjectivity(tweet):
   return TextBlob(tweet).sentiment.subjectivity
```

```python
textblob=pd.DataFrame()
textblob['Tweet']=df['Tweets']
textblob['Polarity']=textblob['Tweet'].apply(getPolarity)
textblob['Subjectivity']=textblob['Tweet'].apply(getSubjectivity)
textblob
```

```python
textblob.to_csv('Adityas TextBlob.csv')
```

```python
pos=[]
neg=[]
neu=[]
for i in textblob['Polarity']:
    if i>(0.2):
        pos.append(i)
    elif i<(-0.2):
        neg.append(i)
    else:
        neu.append(i)
print(len(pos))
print(len(neg))
print(len(neu))
```

```python
from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer
sid_obj= SentimentIntensityAnalyzer()
```

```python
print(sid_obj.polarity_scores("I love my mom")) 
```

```python
def pos(tweet):
    return sid_obj.polarity_scores(tweet)['pos']
def neu(tweet):
    return sid_obj.polarity_scores(tweet)['neu']
def neg(tweet):
    return sid_obj.polarity_scores(tweet)['neg']
def comp(tweet):
    return sid_obj.polarity_scores(tweet)['compound']
```

```python
vader_df=pd.DataFrame()
vader_df['Tweet']=df['Tweets']
vader_df['Positive']=vader_df['Tweet'].apply(pos)
vader_df['Neutral']=vader_df['Tweet'].apply(neu)
vader_df['Negative']=vader_df['Tweet'].apply(neg)
vader_df['Compound']=vader_df['Tweet'].apply(comp)
vader_df
```

```python
vader_df.to_csv('Adityas Vader.csv')
```

```python
posv=[]
negv=[]
neuv=[]
for i in vader_df['Compound']:
    if i>(0.2):
        posv.append(i)
    elif i<(-0.2):
        negv.append(i)
    else:
        neuv.append(i)
print(len(posv))
print(len(negv))
print(len(neuv))
```
