```python
# importing the required module
import numpy as np
from scipy.interpolate import make_interp_spline
import matplotlib.pyplot as plt
 



y = np.array([0,31562,40960,27205,32442,23713,35611,34859,38756,0])

x= [' Mar 24 23:30','Mar 25 11:30','Mar 25 23:30','Mar 26 11:30','Mar 26 23:30','Mar 27 12:30','Mar 28 00:30','Mar 28 12:30','Mar 29 00:30','Mar 29 12:30']

xf=np.arange((len(x)))
X_Y_Spline = make_interp_spline(xf, y)
 
# Returns evenly spaced numbers
# over a specified interval.
X_ = np.linspace(xf.min(), xf.max(), 500)
Y_ = X_Y_Spline(X_)
 
# plotting the points

plt.plot(X_,Y_, color= "red",linestyle="solid")
plt.gca().yaxis.set_major_formatter(plt.matplotlib.ticker.StrMethodFormatter('{x:,.0f}'))
plt.xticks(xf,x,rotation=90)

plt.xlabel('Time (GMT)')
# naming the y axis
plt.ylabel('Number of Tweets')

#plt.title('Distribution of Tweets over Retreival')

plt.grid()
plt.figure(figsize=(20,15))
#plt.savefig('plot.png', dpi=300)
plt.show()
```
